```python
import pickle
import matplotlib.pyplot as plt

def avg(lst):
    return sum(lst)/len(lst)

time_idx = [3,6,9,12,15,18,21,24]

results = [
    [],
    []
]

for idx in range(8):
    time_window = time_idx[idx]
    print("###########################################################")
    result_0 = pickle.load(open(f'./results/REFIT_hack/PEH2_{time_window}h.pkl','rb'))
    print(result_0[7])
    results[0].append(result_0[7][0])
    
for idx in range(8):
    time_window = time_idx[idx]
    print("###########################################################")
    result_0 = pickle.load(open(f'./results/REFIT_hack/VGG19_1D_{time_window}h.pkl','rb'))
    print(result_0[7])
    results[1].append(result_0[7][0])
print("######################################################")

print(results)


fig, ax = plt.subplots()

x = time_idx

ax.plot(x, results[0], label='PEH2')
ax.plot(x, results[1], label='VGG19_1D')

# Set x-axis tick positions and labels
tick_positions = [3, 6, 9, 12, 15, 18, 21, 24]
tick_labels = ['3h', '6h', '9h', '12h', '15h', '18h', '21h', '24h']

ax.set_xticks(tick_positions)
ax.set_xticklabels(tick_labels)
ax.set_xlabel('Time window')

# Set y-axis label
ax.set_ylabel('Weighted F1 score [%]')

# Add legend
ax.legend()

# Display plot
plt.show()



print("Average PEH2: ", round(avg(results[0])*100,1), "%")
print("Average VGG_19D: ", round(avg(results[1])*100,1), "%")
```
